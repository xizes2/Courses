#### Continua amb el conjunt de dades de tema esportiu que t'agradi i ~normalitza~ standaritza les dades tenint en compte els outliers.

```python
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#Boxplot to check outliers
plt.boxplot(df_dh_men['Age'])
```

```python
df_dh_men['Age'].describe()
```

```python
#Determine the low and high quantiles values that will be used to exclude all values below and above these ones.
lower_quant = df_dh_men['Age'].quantile(0.05)
higher_quant = df_dh_men['Age'].quantile(0.95)
```

```python
#Rows with values below the lower quantile. These will not be taken into consideration
df_dh_men[df_dh_men['Age'] < lower_quant]
```

```python
#Rows with values above the higher quantile. These will not be taken into consideration
df_dh_men[df_dh_men['Age'] > higher_quant]
```

```python
#Create df without the outliers
df_no_outliers = df_dh_men[(df_dh_men['Age'] > lower_quant) & (df_dh_men['Age'] < higher_quant)]
```

```python
#Boxplot without outliers
plt.boxplot(df_no_outliers['Age'])
```

```python
#Create Scaler object
scaler_no_outliers = StandardScaler()
```

```python
#Standardize the 'Age' column
scaler_no_outliers.fit_transform(np.array(df_no_outliers['Age']).reshape(-1, 1))
```
