## Exercici 2
Connecta la base de dades NoSQL a Python utilitzant per exemple pymongo.

```python
# Install and import pymongo library
#!pip install pymongo
import pymongo
```

```python
# Create conection with mongo database
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

# Conect with my database
mydb = myclient["hotel_mongodb"]
```

```python
# View all the collections in my database
mydb.list_collection_names()
```

```python
# Create collections object
clients = mydb.get_collection('clients')
```

```python
# Print all documents in collection
for i in clients.find():
    print(i)
    print()
```

# Nivell 2

## Exercici 1
Carega algunes consultes senzilles a un Pandas Dataframe. 

```python
import pandas as pd
```

```python
# Print specific keys of the collection
for i in clients.find({},{'_id':0, 'name':1, 'ck_in':1, 'ck_out':1}):
    try:
        print(i['name'],'->','in:',i['ck_in'].date(),'out:',i['ck_out'].date())
    except:
        print(i['name'],':', 'no information')
        pass
```

```python
# Create dataframe with all the information from the collection
df1 = pd.DataFrame(clients.find({}))
df1
```

```python
df2 = pd.DataFrame(clients.find({}, {'_id':0, 'name':1, 'ck_in':1, 'ck_out':1}))
df2
```

```python
df3 = pd.DataFrame(clients.find({}, {'_id':0, 'name':1, 'email':1, 'document':1}))
df3
```
