#### Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe. Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

```python
import pandas as pd
```

```python
df_delay = pd.read_csv('C:/Users/xizes/DelayedFlights.csv')
```

```python
df_delay.head()
```

```python
df_delay.drop(['Unnamed: 0', 'FlightNum', 'TailNum', 'DayofMonth', 'DayOfWeek', 'TaxiIn', 'TaxiOut', 'Diverted'], axis=1, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_delay.head()
```

```python
#Rearrange columns order
cols = df_delay.columns.to_list()
cols = ['Year', 'Month', 'UniqueCarrier', 'Origin', 'Dest', 'Distance', 'CRSDepTime', 'DepTime', 'DepDelay', 'AirTime', 'CRSArrTime', 'ArrTime', 'ArrDelay', 'ActualElapsedTime', 'Cancelled', 'CancellationCode', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']

df_delay = df_delay[cols]

df_delay.head()
```

```python
print(f'shape: {df_delay.shape}')
print(f'size: {df_delay.size}')
```

```python
#Describe function for delay columns (using slicing with step=2 because of the df´s size)
df_delay[::2].describe()
```

```python
#Total amount of nan values on the dataset
df_delay.isnull().sum()
```

```python
#We chose to 'drop' nan values using 'ArrDelay' as reference because it´s a small number of observations
df_delay = df_delay[df_delay['ArrDelay'].notna()]
```

```python
df_delay.shape
```

```python
df_delay.isnull().sum()
```

```python
#Create Average Fligh Speed (mi/h) column
df_delay['AvgFlightSpeed'] = df_delay['Distance']/(df_delay['ActualElapsedTime']/60)
```

```python
df_delay.head(100)
```

```python
#Amount of companies on the dataset
len(df_delay['UniqueCarrier'].unique())
```

```python
#Total amount of flights
total_flights = df_delay.groupby('UniqueCarrier')['DepDelay'].count().sort_values()

#Proportional total departure delay per company
total_delays_hours = df_delay.groupby('UniqueCarrier')['DepDelay'].sum().sort_values()

print(f'Total amount of hours delayed on departure per carrier: {total_delays_hours/60}. \n\nMinutes delayed on departure per flight per carrier: {(total_delays_hours/total_flights).sort_values()}')
```

```python
#Mean departure delay per airport
df_delay.groupby(['Origin'])['DepDelay'].mean().sort_values(ascending=False)
```

```python
#Mean arrival delay per airport
df_delay.groupby('Dest')['ArrDelay'].mean().sort_values(ascending=False)
```

```python
#Mean arrival delay per month
df_delay.groupby('Month')['ArrDelay'].mean().sort_values(ascending=False)
```

```python
#Configure the float format display from scietific notation to 3 decimals
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
#Statistical data about the arrival delays
df_delay['ArrDelay'].describe()
```

```python
#'Legs' with the highest mean arrival delay
df_delay.groupby(['Origin', 'Dest'])['ArrDelay'].mean().sort_values(ascending=False).head(10)
```

```python
#Three longest flights by distance
df_delay[['Origin', 'Dest', 'Distance']].sort_values('Distance', ascending=False).head(1000).drop_duplicates(['Distance'])
```

```python
#Five most delayed flights
df_delay[['Origin', 'Dest', 'ArrDelay']].sort_values('ArrDelay', ascending=False).head()
```
